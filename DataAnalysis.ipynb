```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Camera_TTCS_Experiments.xlsx")
df.head()
```

```python
df.columns
```

```python
SHITOMASI = ['SHITOMASI_BRISK', 'SHITOMASI_BRIEF', 'SHITOMASI_FREAK',
       'SHITOMASI_ORB', 'SHITOMASI_SIFT']
HARRIS = ['HARRIS_BRISK', 'HARRIS_BRIEF','HARRIS_FREAK', 'HARRIS_ORB', 'HARRIS_SIFT']

BRISK = ['BRISK_BRISK',
       'BRISK_BRIEF', 'BRISK_FREAK', 'BRISK_ORB', 'BRISK_SIFT']

ORB = ['ORB_BRISK',
       'ORB_BRIEF', 'ORB_FREAK', 'ORB_ORB', 'ORB_SIFT']

FAST = ['FAST_BRISK', 'FAST_BRIEF', 'FAST_FREAK', 'FAST_ORB', 'FAST_SIFT']

AKAZE = ['AKAZE_AKAZE']
SIFT = ['SIFT_BRISK', 'SIFT_BRIEF', 'SIFT_FREAK', 'SIFT_SIFT']
```

```python
def plotTTC(df,lists,listName):
  ax = df[lists].plot(figsize = (15,6))
  ax.set_xticks(df.index)
  ax.set_title(listName+" Detector")
  ax.set_xlabel("Frame")
  ax.set_ylabel("Camera Based TTC in s")
  plt.ylim((-1,50))
  plt.grid()
  plt.savefig(listName+"_Dector.png",dpi=300)
```

```python
plotTTC(df,SHITOMASI,"SHITOMASI")
```

```python
plotTTC(df,HARRIS,"HARRIS")
```

```python
plotTTC(df,BRISK,"BRISK")
```

```python
plotTTC(df,ORB,"ORB")
```

```python
plotTTC(df,FAST,"FAST")
```

```python
plotTTC(df,AKAZE,"AKZE")
```

```python
plotTTC(df,SIFT,"SIFT")
```
